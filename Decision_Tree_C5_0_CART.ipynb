```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import datasets
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
```

```python
# import some data to play with
iris = pd.read_csv('https://raw.githubusercontent.com/slmsshk/DataSet/main/iris.csv')
```

```python
iris.head()
```

```python
iris['Species'].value_counts()
```

```python
#Complete Iris dataset
label_encoder = preprocessing.LabelEncoder()
iris['species']= label_encoder.fit_transform(iris['Species'])
```

```python
x=iris.iloc[:,0:4]
y=iris['Species']
```

```python
iris['Species'].unique()
```

```python
y=iris['Species']
```

```python
iris['Species'].value_counts()
```

```python
iris.columns
```

```python
colnames = list(iris.columns)
```

```python
# Splitting data into training and testing data set
# from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=40)
```

# Building Decision Tree Classifier using Entropy Criteria

```python
model = DecisionTreeClassifier(criterion = 'gini', max_depth=3)
model.fit(x_train,y_train) 
```

```python
model
```

```python
#Plot the decision tree
from sklearn import tree
tree.plot_tree(model);
```

```python
y_train.value_counts().keys()
```

```python
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'] #.columns
cn=['setosa', 'versicolor', 'virginica']           #.value_count.keys
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4),  dpi=500)
tree.plot_tree(model,
               feature_names = fn,
               class_names=cn,
               filled = True);
```

```python
#Predicting on test data
preds = model.predict(x_test)  # predicting on test data set
pd.Series(preds).value_counts()  # getting the count of each category
```

```python
y_test.value_counts()
```

```python
preds
```

```python
pd.crosstab(y_test,preds)  # getting the 2 way table to understand the correct and wrong predictions
```

```python
preds,y_test
```

```python
# Accuracy
from sklearn.metrics import accuracy_score

np.mean(preds==y_test),accuracy_score(y_test,preds)
```
