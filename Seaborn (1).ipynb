# **Seaborn**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips =sns.load_dataset('tips')
```

```python
tips.head()
```

# univariate--->"One variable"data visualization

```python
#strip plot
sns.stripplot(x='sex',hue='day',data=tips,jitter=True)
plt.title('strip plot')
plt.xlabel('Gender')
plt.show()
```

```python
#Grouping with stripplot()
sns.stripplot(x='sex', y='tip',hue='day', data=tips)
plt.ylabel('tips($)')
plt.show()
```

```python
sns.stripplot(x='day',y='tip', data=tips, size=4,jitter=False)
plt.ylabel('tip ($)')
plt.show
```

```python
#Swarm plot
plt.figure(figsize=(10,5))
sns.swarmplot(x='day', y='tip', data=tips,facecolor='black')
sns.violinplot(x='day',y='tip',data=tips,color='black')
plt.ylabel('tip($)')
plt.show
```

```python
sns.swarmplot(x='day',y='tip', data=tips, hue='sex')
plt.ylabel('tip($)')
plt.show()
```

```python
#Changing orientation
sns.swarmplot(x='tip', y='day', data=tips, hue='sex',orient='h')
plt.xlabel('tip ($)')
plt.show()
```

# Box and Violin plot

```python
# plt.subplot(1,2,1)
plt.figure(figsize=(10,10))
sns.boxenplot(x='day' , y='tip', data=tips,)
sns.violinplot(x='day', y='tip',data=tips,color='lightgrey')

plt.ylabel('tip($)')
plt.show
```

```python
#plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')

plt.show()
```

```python
plt.subplots(nrows=2,ncols=2)
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')

plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip($)')

plt.show()
```

```python
## combing plots

sns.violinplot(x='day', y='tip', data=tips, inner=None,color='lightgray')
sns.stripplot(x='day', y='tip', data=tips, size=4,jitter=True)
plt.ylabel('tip($)')
plt.show()
```

# Bivarite --> "Two variables" data **visualization**

```python
#joint plot
sns.jointplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
# Density plot
sns.jointplot(x='total_bill', y='tip', data=tips,kind='kde')
plt.show()
#contours are created representing various levels of density of the data
```

```python
# pair plot
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, hue='sex')
plt.show()
```

# **Titanic data visualization                                   **

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

# Loading dataset

```python
data = pd.read_csv("https://raw.githubusercontent.com/slmsshk/DataSet/main/TT(train).csv")
data.head()
```

# **survival-Survival(0=No; 1 = Yes)**


class - Passenger Class (1 = 1st; 2 = 2nd; 3 = 3rd) name - Name sex - Sex age - Age sibsp - Number of Siblings/Spouses Aboard parch - Number of Parents/Children Aboard ticket - Ticket Number fare - Passenger Fare cabin - Cabin embarked - Port of Embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)


# **Data** **types**

```python
data.dtypes
```

# **proportion of target (Survived)**

```python
data.groupby('Survived')['PassengerId'].count()
```

# **Visual Exploration**

```python
# f, ax = plt.subplots(figsize=(11,5))
plt.figure(figsize=(10,5))
sns.boxplot(x='Survived', y="Age", data=data);
```

```python
# f, ax = plt.subplots(figsize=(10,8))
plt.figure(figsize=(10,8))
sns.barplot(x='Sex', y="Survived", data=data);
```

```python
# f, ax = plt.subplots(figsize=(11,5))
plt.figure(figsize=(11,5))
sns.boxplot(x="Sex", y="Age", hue="Survived", data=data);
```

```python
sns.barplot(x="Pclass", y="Survived", data=data);
```

```python
sns.barplot(x="Pclass", y="Survived", hue="Sex", data=data);
```

```python
sns.barplot(x="SibSp", y="Survived", data=data);
```

```python
sns.barplot(x="Parch", y="Survived", data=data);
```

```python
survived = data.loc[data['Survived']==1,"Age"].dropna()
sns.distplot(survived)
plt.title("Survived")
```

```python
not_survived = data.loc[data['Survived']==0,"Age"].dropna()
sns.distplot(not_survived)
plt.title("Not Survived");
```

```python
sns.pairplot(data.dropna());
```

```python
# Pclass vs Survive
grid = sns.FacetGrid(data, col='Survived', row='Pclass', size=2.4, aspect=1.5)
grid.map(plt.hist, 'Age', alpha=.5, bins=20)
grid.add_legend()
```

```python
sns.displot(a=data[data['Embarked']=='C']['Survived'],bins=3,kde=False)
plt.title("Cherbourg")
plt.xticks([0,1])
plt.show()
plt.title("QueensTown")
sns.distplot(a=data[data['Embarked']=='Q']['Survived'],bins=3,kde=False)
plt.xticks([0,1])

plt.show()
plt.title("Southampton")
sns.displot(a=data[data['Embarked']=='S']['Survived'],bins=3,kde=False)
plt.xticks([0,1])

plt.show()
# most of the Passengers embarked from Southampton
```

```python
figbi,axesbi = plt.subplots(2,4,figsize=(18, 10))
data.groupby('Pclass')['Survived'].mean().plot(kind='barh',ax=axesbi[0,0],xlim=[0,1])
data.groupby('SibSp')['Survived'].mean().plot(kind='barh',ax=axesbi[0,1],xlim=[0,1])
data.groupby('Parch')['Survived'].mean().plot(kind='barh',ax=axesbi[0,2],xlim=[0,1])
data.groupby('Sex')['Survived'].mean().plot(kind='barh',ax=axesbi[0,3],xlim=[0,1])
data.groupby('Embarked')['Survived'].mean().plot(kind='barh',ax=axesbi[1,0],xlim=[0,1])
sns.boxplot(x="Survived", y="Age", data=data,ax=axesbi[1,1])
sns.boxplot(x="Survived", y="Fare", data=data,ax=axesbi[1,2])
plt.show()
```
